```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
# Functional
plt.plot(x,y)

# use plt.show() to show outside jupyter notebook
plt.xlabel('x')
plt.ylabel('y')
plt.title('titl')
```

```python
# multiplots

plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python

#OO

fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_title('titl')
axes.set_ylabel('y')
```

```python
fig= plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Big')
axes2.plot(y,x)
axes2.set_title('Small')
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,2))
for ax in axes:
    ax.plot(x,y**2,label='X**2',color='purple',lw=3,alpha=0.5,ls='-',marker='o',ms=2)
    ax.legend(loc=0)
    ax.set_xlim([0,2])
#     ax.set_title(ax)
# axes.plot(x,y)
plt.tight_layout()
```

```python
fig.savefig('pic.jpg')
```
